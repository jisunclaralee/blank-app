1. # 맷플롯립의 구조
##1.1 파이플롯

```python
import matplotlib.pyplot as plt

X = range(100)
y = range(100)
plt.plot(X, y)
plt.show()
```

```python
import numpy as np

X_1 = np.arange(100)
Y_1 = [np.cos(value) for value in X]

X_2 = np.arange(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

## 1.2 그림과 축

```python
fig, ax = plt.subplots()

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

ax.plot(X_1, Y_1)
ax.set(title="cos graph",
             xlabel='X',
             ylabel='Y')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c="b")
ax_2.plot(X_1, Y_1, c="g")
plt.show()
```

# 2.맷플롯립으로 그래프 꾸미기
## 1.색상

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value +100 for value in X]

plt.plot(X_1, Y_1, color = '#000000')
plt.plot(X_2, Y_2, c = 'c')

plt.show()
```

## 2.2선의 형태

```python
plt.plot(X_1, Y_1, c='b', linestyle = 'dashed')
plt.plot(X_2, Y_2, c='r', ls = 'dotted')

plt.title("Two lines")
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax_1=fig.add_subplot(1,2,1)
ax_2=fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c='b')
ax_1.set_title('figure 1')
ax_2.plot(X_2, Y_2, c='g')
ax_2.set_title('figure 2')

plt.show()
```

## 2.4 범례

```python
plt.plot(X_1, Y_1,
         color='b',
         linestyle = 'dashed',
         label = 'line_1')
plt.plot(X_2, Y_2,
         color='r',
         linestyle = 'dotted',
         label = 'line_2')
plt.legend(
    shadow=True,
    fancybox=False,
    loc='upper right'
)

plt.title('$y =ax+b$')
plt.xlabel('$x_line$')
plt.show()
```

#3. 맷플롯립에서 사용하는 그래프
## 1.산점도

```python
data_1 = np.random.rand(512,2)
data_2 = np.random.rand(512,2)

plt.scatter(data_1[:,0],
            data_1[:,1],
            c='b', marker='x')
plt.scatter(data_2[:,0],
            data_2[:,1],
            c='r', marker='o')
plt.show()
```

```python
N =50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (
    15 * np.random.rand(N))**2

plt.scatter(x,y,
              s=area, c=colors,
              alpha=0.5)
plt.show()
```

##3.2 막대그래프

```python
data = [[5., 25., 50., 20.],
        [4.,23.,51.,17],
        [6.,22.,52.,19]]

X = np.arange(0,8,2)

plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)

plt.xticks(X+0.50, ('A','B','C','D'))

plt.show()
```

## 3.3누적 막대그래프

```python
data = np.array([[5., 25., 50., 20.],
 [4., 23., 51., 17],
  [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])

data = np.array([[5., 5., 5., 5.],
 [4., 23., 51., 17],
[6., 22., 52., 19]])

for i in range(3):
  plt.bar(X,data[i],
          bottom = np.sum(
              data[:i],axis=0),
          color = color_list[i],
          label = data_label[i])
plt.legend()
plt.show()
```
